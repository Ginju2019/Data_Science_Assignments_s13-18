# 1. A test is conducted which is consisting of 20 MCQs (multiple choices questions) with every MCQ having its four options out of which only one is correct. Determine the probability that a person undertaking that test has answered exactly 5 questions wrong.

__Solution:__

```python
#n=20
#4 option, so 3 are wrong and 1 is correct
#Probability of correct answers from 1 question=1/4=0.25
#Probability of incorrect answers from 1 question=3/4=.75
#5 wrong means 15 correct
#incorrect answer from 5 question= 20*3/4
#find 5 wrong from 20
```

```python
P5Qs_wrong= binom.pmf(5,20, 0.75)#probability mass function for binom
print("Probability that a person undertaking 20 MCQ test has answered exactly 5 questions wrong: ", P5Qs_wrong)
```

##### Also can be done as:

```python
def prob_binomial(p,q,n,r): # user defined function
    prob=scipy.special.comb(n,r)*(p**r)*(q**(n-r))
    return prob

P5Qs_wrong = prob_binomial((3/4),(1/4),20,5) # calling user defined function prob_binomial
print("Probability that a person undertaking 20 MCQ test has answered exactly 5 questions wrong:",P5Qs_wrong)
```
